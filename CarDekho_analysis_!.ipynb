```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
cardekho_datasets_train, cardekho_datasets_test = train_test_split(cardekho_datasets, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() #helps to balance the min and max values in training and testing data sets
```

```python
num_vars = ['km_driven', 'seats', 'selling_price','year']

cardekho_datasets_train[num_vars] = scaler.fit_transform(cardekho_datasets_train[num_vars])
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Features'] = x_train_lm.columns
vif['VIF'] = [variance_inflation_factor(x_train_lm.values, i) for i in range(x_train_lm.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
num_vars = ['km_driven', 'seats', 'selling_price','year']

cardekho_datasets_test[num_vars] = scaler.fit_transform(cardekho_datasets_test[num_vars])
```

```python
y_test = cardekho_datasets_test.pop('selling_price')
X_test = cardekho_datasets_test
```

```python
X_test_m4 = sm.add_constant(X_test)
```

```python
X_test_m4.columns
```

```python
X_test_m4 = X_test_m4.drop([
        'CNG', 'LPG', 'Petrol',
       'First Owner', 'Fourth & Above Owner', 'Second Owner',
       'Third Owner'], axis = 1)
```

```python
X_test_m4.columns
```

```python
y_pred_m4 = lr.predict(X_test_m4.astype(float))
```

```python
fig = plt.figure()
plt.scatter(y_test, y_pred_m4)
fig.suptitle('y_test vs y_pred', fontsize = 20)              # Plot heading 
plt.xlabel('y_test', fontsize = 18)                          # X-label
plt.ylabel('y_pred', fontsize = 16)   
m, b = np.polyfit(y_test, y_pred_m4, 1)
plt.plot(y_test, m*y_test + b)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))
ind = np.arange(15)
width = 0.35

plt.bar(ind, y_test[:15], width, label='actual selling price',color=['Blue'])
plt.bar(ind, y_pred_m4[:15], width, label='predicted_selling_price',color=['green'])



plt.xlabel("cars")
plt.ylabel("selling price")
plt.title("Selling Price - Actual vs Predicted")

plt.show()
```
